# Plotting with Matplotlib

## Anatomía de un gráfico de Matplotlib

````python
# This figure shows the name of several matplotlib elements composing a figure

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator, FuncFormatter


np.random.seed(19680801)

X = np.linspace(0.5, 3.5, 100)
Y1 = 3+np.cos(X)
Y2 = 1+np.cos(1+X/0.75)/2
Y3 = np.random.uniform(Y1, Y2, len(X))

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, aspect=1)


def minor_tick(x, pos):
    if not x % 1.0:
        return ""
    return "%.2f" % x

ax.xaxis.set_major_locator(MultipleLocator(1.000))
ax.xaxis.set_minor_locator(AutoMinorLocator(4))
ax.yaxis.set_major_locator(MultipleLocator(1.000))
ax.yaxis.set_minor_locator(AutoMinorLocator(4))
ax.xaxis.set_minor_formatter(FuncFormatter(minor_tick))

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

ax.tick_params(which='major', width=1.0)
ax.tick_params(which='major', length=10)
ax.tick_params(which='minor', width=1.0, labelsize=10)
ax.tick_params(which='minor', length=5, labelsize=10, labelcolor='0.25')

ax.grid(linestyle="--", linewidth=0.5, color='.25', zorder=-10)

ax.plot(X, Y1, c=(0.25, 0.25, 1.00), lw=2, label="Blue signal", zorder=10)
ax.plot(X, Y2, c=(1.00, 0.25, 0.25), lw=2, label="Red signal")
ax.plot(X, Y3, linewidth=0,
        marker='o', markerfacecolor='w', markeredgecolor='k')

ax.set_title("Anatomy of a figure", fontsize=20, verticalalignment='bottom')
ax.set_xlabel("X axis label")
ax.set_ylabel("Y axis label")

ax.legend()


def circle(x, y, radius=0.15):
    from matplotlib.patches import Circle
    from matplotlib.patheffects import withStroke
    circle = Circle((x, y), radius, clip_on=False, zorder=10, linewidth=1,
                    edgecolor='black', facecolor=(0, 0, 0, .0125),
                    path_effects=[withStroke(linewidth=5, foreground='w')])
    ax.add_artist(circle)


def text(x, y, text):
    ax.text(x, y, text, backgroundcolor="white",
            ha='center', va='top', weight='bold', color='blue')


# Minor tick
circle(0.50, -0.10)
text(0.50, -0.32, "Minor tick label")

# Major tick
circle(-0.03, 4.00)
text(0.03, 3.80, "Major tick")

# Minor tick
circle(0.00, 3.50)
text(0.00, 3.30, "Minor tick")

# Major tick label
circle(-0.15, 3.00)
text(-0.15, 2.80, "Major tick label")

# X Label
circle(1.80, -0.27)
text(1.80, -0.45, "X axis label")

# Y Label
circle(-0.27, 1.80)
text(-0.27, 1.6, "Y axis label")

# Title
circle(1.60, 4.13)
text(1.60, 3.93, "Title")

# Blue plot
circle(1.75, 2.80)
text(1.75, 2.60, "Line\n(line plot)")

# Red plot
circle(1.20, 0.60)
text(1.20, 0.40, "Line\n(line plot)")

# Scatter plot
circle(3.20, 1.75)
text(3.20, 1.55, "Markers\n(scatter plot)")

# Grid
circle(3.00, 3.00)
text(3.00, 2.80, "Grid")

# Legend
circle(3.70, 3.80)
text(3.70, 3.60, "Legend")

# Axes
circle(0.5, 0.5)
text(0.5, 0.3, "Axes")

# Figure
circle(-0.3, 0.65)
text(-0.3, 0.45, "Figure")

color = 'blue'
ax.annotate('Spines', xy=(4.0, 0.35), xycoords='data',
            xytext=(3.3, 0.5), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.annotate('', xy=(3.15, 0.0), xycoords='data',
            xytext=(3.45, 0.45), textcoords='data',
            weight='bold', color=color,
            arrowprops=dict(arrowstyle='->',
                            connectionstyle="arc3",
                            color=color))

ax.text(4.0, -0.4, "Made with http://matplotlib.org",
        fontsize=10, ha="right", color='.5')

plt.show()
````

$\textbf{Componentes principales de la "anatomía de un gráfico de Matplotlib:}$

$\textbf{Figure:}$ Es un ojeto de matplotlib que permite el despliegue de gráficos obtenidos a partir de ciertos datos.

$\textbf{Axes:}$ Literalmente traduce "ejes" y se refiere a un plano de coordenadas individual donde se plasma una gráfica en un ojeto "Figure".

$\textbf{Title:}$ Es el título que se utiliza para distinguir la gráfica a través de una denominación descriptiva.

$\textbf{Y axis label:}$ Es el texto que describe los datos del eje de ordenadas (variable dependiente).

$\textbf{X axis label:}$ Es el texto que describe los datos del eje de abscisas (variable independiente).

$\textbf{Line:}$ Es el trazo que conecta los puntos de pares de coordenadas en el plano de coordenadas de un gráfico. Puede ser un trazo continuo o discontinuo dependiendo del estilo que se escoja para el mismo (lines.lineStyles.keys() -> ['', ' ', 'None', '--', '-.', '-', ':']). 

$\textbf{Marker:}$ Son marcas que se utilizan para señalar pares de coordenadas individuales en el plano. Se utilizan especialmente en los gráficos de tipo "scatter plot". Matplotlib permite escoger el estilo de los marcadores. (markers = {'.': 'point', ',': 'pixel', 'o': 'circle', 'v': 'triangle_down', '^': 'triangle_up', '<': 'triangle_left', '>': 'triangle_right', '1': 'tri_down', '2': 'tri_up', '3': 'tri_left', '4': 'tri_right', '8': 'octagon', 's': 'square', 'p': 'pentagon', '*': 'star', 'h': 'hexagon1', 'H': 'hexagon2', '+': 'plus', 'x': 'x', 'D': 'diamond', 'd': 'thin_diamond', '|': 'vline', '_': 'hline', 'P': 'plus_filled', 'X': 'x_filled', 0: 'tickleft', 1: 'tickright', 2: 'tickup', 3: 'tickdown', 4: 'caretleft', 5: 'caretright', 6: 'caretup', 7: 'caretdown', 8: 'caretleftbase', 9: 'caretrightbase', 10: 'caretupbase', 11: 'caretdownbase', 'None': 'nothing', None: 'nothing', ' ': 'nothing', '': 'nothing'})

$\textbf{Thicks (major y minor):}$ Son marcas (o valores) en los ejes de coordenadas que señalan puntos específicos en los ejes de coordenadas.

$\textbf{Legend:}$ Texto acompañado de un segmento de línea (conservando el estilo aplicado a la línea) que permite distinguir varias gráficas que están dibujadas en un mismo eje de coordenadas.

$\textbf{Grid:}$ Propiedad que permite la visualización de la cuadrícula en el sistema de coordenadas.

$\textbf{Spines:}$ Son básicamente los bordes que delimitan las gráficas.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,10,1)
y1 = np.ones(len(x)) 
y2 = 2*np.ones(len(x)) 
y3 = 3*np.ones(len(x)) 
y4 = 4*np.ones(len(x)) 
y5 = 5*np.ones(len(x))
y6 = 6*np.ones(len(x))
y7 = 7*np.ones(len(x))
y8 = 8*np.ones(len(x))
y9 = 9*np.ones(len(x))
y10 = 10*np.ones(len(x))
y11 = 11*np.ones(len(x))
y12 = 12*np.ones(len(x))
y13 = 13*np.ones(len(x))
y14 = 14*np.ones(len(x))
y15 = 15*np.ones(len(x))
y16 = 16*np.ones(len(x))
# Diferentes colores y estilos de trazo para líneas. Diferentes colores y símbolos para marcadores
plt.plot(x, y1, 'k')  # Negro, línea continua
plt.plot(x, y2, 'r--') # Rojo, línea discontinua guiones
plt.plot(x, y3, 'g-.') # Verde, línea discontinua guión-punto
plt.plot(x, y4, 'b-') # Azul, línea continua
plt.plot(x, y5, 'y:') # Amarillo, línea discontinua puntos
plt.plot(x, y6, 'c.') # Cyan, marcador punto
plt.plot(x, y7, 'm*') # Magenta, marcador estrella
plt.plot(x, y8, 'k^') # Negro, marcador triangulo arriba
plt.plot(x, y9, 'r+') # Rojo, marcador signo más
plt.plot(x, y10, 'go') # Verde, marcador circulo
plt.plot(x, y11, 'b<') # Azul, marcador triangulo izquierda
plt.plot(x, y12, 'y>') # Amarillo, marcador triangulo derecha
plt.plot(x, y13, 'cs') # Cyan, marcador cuadrado
plt.plot(x, y14, 'mp') # Magenta, marcador pentágono
# Combinando líneas y marcadores
plt.plot(x, y15, 'k', y15, 'k^') # Línea continua, marcador triangulo arriba
plt.plot(x, y16, 'r--', y16, 'r+') # Línea discontinua, marcador signo más
plt.show()
````

$\textbf{Propiedades de uso común en los gráficos de Matplotlib:}$

El método plot de matplotlib.pyplot recibe un número variable de argumentos, de acuerdo con la definición del método:

```python
plot([x], y, [fmt], data=None, **kwargs)
plot([x], y, [fmt], [x2], y2, [fmt2], ..., **kwargs) 
```
A continuación se listan algunas de las propiedades de uso más frecuente para el formato de gráficas en matplotlib:

$\textbf{xdata, ydata:}$ Son los valores de los pares ordenados de la función que se quiere graficar. Son por defecto arreglos de numpy, pero pueden ser filas de un dataframe de pandas, listas (o list comprehensions), funciones.

$\textbf{color:}$ Determina el color de línea del gráfico.

$\textbf{linewidth:}$ Determina el grosor de la línea del gráfico.

$\textbf{linestyle:}$ Determina el estilo de trazo del gráfico.

$\textbf{markersize:}$ Determina el tamaño de los marcadores.

$\textbf{markerfacecolor:}$ Determina el color de relleno de los marcadores.

$\textbf{markeredgecolor:}$ Determina el color del delineado de los marcadores.

$\textbf{label:}$ Agrega una etiqueta al gráfico para que pueda ser luego referenciado con una leyenda.

$\textbf{scalex, scaley:}$ (bool, optional) Determina si los límites del gráfico se adaptan a los límites de los datos.

$\textbf{zorder:}$ Determina el orden de un elemento dentro de un gráfico, es decir, si va debajo o por encima de otro elemento que coincide en la misma ubicación.


````python
import matplotlib.pyplot as plt
import numpy as np

#Vector de abscisas
x = np.arange(0,6,0.01) 
#Graficando x contra x^2 usando linea continua de color azul y asignando una etiqueta a la gráfica
plt.plot(x, [i**2 for i in x], 'b-', label="$x^{2}$") 
#Definiendo los limites de los ejes coordenados
plt.axis([0, 6, 0, 40]) 
#Agregando título a la gráfica
plt.title('Gráfica de x v.s. $x^{2}$')
plt.xlabel("x")
plt.ylabel("$x^{2}$")
#Agregando la etiqueta de la gráfica en un recuadro de leyendas
plt.legend()
#Desplegando la gráfica
plt.show()  
````

````python
# Vector de abscisas
t = np.arange(0., 5., 0.2)
# Varias graficas en un mismo sistema de coordenadas. Se diferencian con distintos estilos de línea
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
# Agregando título a la gráfica
plt.title("Varias gráficas en un mismo plano de coordenadas")
# Rotulando el eje de abscisas
plt.xlabel("t")
# Rotulando el eje de ordenadas
plt.ylabel("t, $t^{2}$, $t^{3}$")
plt.show()
````

Otra forma de graficar varias funciones al tiempo en el mismo sistema de coordenadas:

````python
# Vector de abscisas
t = np.arange(0., 5., 0.2)
a = t
b = t**2
c = t**3
# Varias graficas en un mismo sistema de coordenadas. Se diferencian con distintos estilos de línea
# a = t
plt.plot(t, a, 'r--', label = "t")
# b = t^2
plt.plot(t, b, 'bs', label = "$t^{2}$")
# c = t^3
plt.plot(t, c, 'g^', label = "$t^{3}$")
# Agregando las etiquetas de las gráficas de las funciones en un recuadro de leyendas
plt.legend()
# Agregando título a la gráfica
plt.title("Varias gráficas en un mismo plano de coordenadas")
# Rotulando el eje de abscisas
plt.xlabel("t")
# Rotulando el eje de ordenadas
plt.ylabel("t, $t^{2}$, $t^{3}$")
# Agregando cuadrícula al sistema de coordenadas (cuadícula de línea discontinua)
plt.grid(linestyle="--")
plt.show()
````

Con el método subplot se puede instanciar varios planos de coordenadas (Axes) en un mismo objeto Figure. Se puede definir un arreglo de varias filas y columnas donde cada una de las celdas corresponde a un sistema de coordenadas individual.

````python
def f(t):  #Function to plot
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)   #Eje de abscisas
t2 = np.arange(0.0, 5.0, 0.02)  #Eje de abscisas

plt.figure(figsize=(10,5))
# Sistema de coordenadas de la fila 1 y la columna 1 (El objeto figure tiene 2 filas y una columna)
ax = plt.subplot(211) 
# Agregando título al primer sistema de ejes de coordenadas
ax.set_title('Sistema subamortiguado')
# Graficando dos funciones al mismo tiempo (circulos azules, línea continua de color negro)
ax.plot(t1, f(t1), 'bo', t2, f(t2), 'k') 
# Sistema de coordenadas de la fila 2 y la columna 1 (El objeto figure tiene 2 filas y una columna)
bx = plt.subplot(212) 
# Agregando título al segundo sistema de ejes de coordenadas
bx.set_title('Funcion coseno')
# Greficando en el segundo sistema de coordenadas
bx.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
````

````python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
# Declarando un objeto figure
fig = plt.figure(figsize=(10,5))
# Declarando un título global a todos los gráficos
fig.suptitle('Multiplot con matplotlib')
# Sistema de coordenadas de la fila 1 y la columna 1 (El objeto figure tiene 2 filas y 2 columnas)
ax1 = fig.add_subplot(221)
ax1.plot(x, y)
# Sistema de coordenadas de la fila 1 y la columna 2 (El objeto figure tiene 2 filas y 2 columnas)
ax2 = fig.add_subplot(222)
ax2.scatter(x, y)
# Sistema de coordenadas de la fila 2 y la columna 1 (El objeto figure tiene 2 filas y 2 columnas)
ax3 = fig.add_subplot(223)
ax3.plot(x, y ** 2)
# Sistema de coordenadas de la fila 2 y la columna 2 (El objeto figure tiene 2 filas y 2 columnas)
ax4 = fig.add_subplot(224)
ax4.scatter(x, y ** 2)
# Titulos para los subplots
ax1.title.set_text('First Plot')   
ax2.title.set_text('Second Plot')
ax3.title.set_text('Third Plot')
ax4.title.set_text('Fourth Plot')
plt.show()
````

$\textbf{Algunos gráficos que se pueden realizar con ayuda de matplotlib:}$

$\textbf{scatter (bubble plot, gráfico de dispersión):}$

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.scatter.html

```python
scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, linewidths=None, verts=None, edgecolors=None, *, data=None, **kwargs)
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo 1 (Tomado de: https://pythonspot.com/matplotlib-scatterplot/)
N = 60
g1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
g2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
g3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))

# Datos para cada uno de los 3 gráficos de dispersión
data = (g1, g2, g3)
# Colores para los marcadores de las gráficas
colors = ("red", "green", "blue")
# Etiquetas para el recuadro de leyendas
groups = ("coffee", "tea", "water")

# Objeto figure
fig = plt.figure(figsize=(13,5))
fig.suptitle('Matplotlib scatter plot')
ax = fig.add_subplot(121)

# Graficando los 3 gráficos de dispersión en el mismo sistema de ejes
for data, color, group in zip(data, colors, groups):
    x, y = data
    # Scatter plot: x(abscisas), y(ordenadas), c(color del marcador), s(tamaño del marcador), edgecolors(color delineado), label(etiqueta gráfica)
    ax.scatter(x, y, alpha=0.8, c=color, edgecolors='none', s=30, label=group)

ax.title.set_text('Ejemplo 1')
plt.legend(loc=2)


# Ejemplo 2 (Tomado de: https://matplotlib.org/gallery/shapes_and_collections/scatter.html#sphx-glr-gallery-shapes-and-collections-scatter-py)

# Fixing random state for reproducibility
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
# Este vector se utiliza para dar un color aleatorio a los marcadores de la gráfica
colors = np.random.rand(N)
# Este vector se utiliza para dar un tamaño variable a los marcadores de la gráfica
area = (30 * np.random.rand(N))**2  

bx = fig.add_subplot(122)
# Scatter plot: El parámetro alpha sirve para dar nivel de transparencia al color de los marcadores
bx.scatter(x, y, s=area, c=colors, alpha=0.5)
bx.title.set_text('Ejemplo 2')
plt.show()
````

$\textbf{Gráfico en coordenadas polares:}$

https://matplotlib.org/api/_as_gen/matplotlib.pyplot.polar.html

```python
polar(theta, r, **kwargs)
```

````python
import numpy as np
import matplotlib.pyplot as plt

# Ejemplo 1 (Tomado de: https://matplotlib.org/examples/pylab_examples/polar_demo.html)

# Vector magnitud
r = np.arange(0, 2, 0.01)
# Vector ángulo
theta = 2 * np.pi * r

fig = plt.figure(figsize=(9,9))
# Indicando que se va a graficar en coordenadas polares
ax = fig.add_subplot(221, projection='polar')
ax.plot(theta, r, 'b-')
ax.set_rmax(2)
ax.set_rticks([0.5, 1, 1.5, 2])  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')

# Ejemplo 2 (Tomado de: https://matplotlib.org/gallery/misc/transoffset.html#sphx-glr-gallery-misc-transoffset-py)
# Indicando que se va a graficar en coordenadas polares
bx = fig.add_subplot(222, projection='polar')
plt.polar(theta, r, 'ro')

# Ejemplo 3 (Tomado de: https://matplotlib.org/gallery/pie_and_polar_charts/polar_scatter.html#sphx-glr-gallery-pie-and-polar-charts-polar-scatter-py)
np.random.seed(19680801)

# Tamaños y colores aleatorios para los marcadores
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta

fig = plt.figure(figsize=(9,9))
ax = fig.add_subplot(223, projection='polar')
# Gráfico de dispersión en coordenadas polares, cmap(mapa de colores, solo se usa cuando colors es un valor en punto flotante)
c = ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75)

plt.show()
````
